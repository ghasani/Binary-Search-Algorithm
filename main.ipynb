```python
random_list = [96, 4, 6, 88, 22, 56, 78, 89]
```

```python
def bin_algo(x):
    sorted_list = sorted(random_list)
    for i in range(len(random_list)):
        h1 = sorted_list[:int(len(sorted_list)/2)]
        h2 = sorted_list[int(len(sorted_list)/2):]
        if x in h1 and len(sorted_list) >= 1:
            sorted_list = h1
        elif x in h2 and len(sorted_list) >= 1:
            sorted_list = h2
    if len(sorted_list) == 1:
        return 'Found'
    else:
        return 'Not Found'
```

```python
bin_algo(89)
```
